```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("E://PHD/Course materials/Sem 1/Data Mining/Assignment/workspace/stock_price_trend_prediction/stock_price_pred_data_mining/Data/01_raw/stock_list_new.csv")
```

```python
df.count()
```

```python
plt.figure(figsize=(4,2))
sns.countplot(data=df, x='market_capitalization', palette='viridis')
plt.title('Stock Distribution by Market Capitalization')
plt.xlabel('Market Capitalization Category')
plt.ylabel('Number of Stocks')
plt.show()
```

```python
plt.figure(figsize=(5,2))
df['sector'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Sectors by Stock Count')
plt.xlabel('Sector')
plt.ylabel('Number of Stocks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=df, x='sector', hue='market_capitalization', palette='Set2')
plt.title('Sector-wise Distribution by Market Capitalization')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Number of Stocks')
plt.show()
```
