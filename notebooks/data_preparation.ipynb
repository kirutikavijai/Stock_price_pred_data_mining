```python
all_data = []

for stock in tqdm(df['stock_id']):
    stock = str(stock).strip().upper() 
    
    data = yf.download(stock + ".NS", start=start_date, end=end_date)
    data['stock_id'] = stock
    print("++++++++++++",start_date)
    print(data)
    all_data.append(data)

# Reset index for clean formatting
all_data = pd.concat(all_data, ignore_index=True)
```
