```python
import yfinance as yf
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("E://PHD/Course materials/Sem 1/Data Mining/Assignment/workspace/stock_price_trend_prediction/stock_price_pred_data_mining/Data/01_raw/stock_list_new.csv")
```

```python
start_date = "2020-01-01"
end_date = "2025-09-30"
```

```python
all_data = []

for stock in tqdm(df['stock_id']):
    stock = str(stock).strip().upper()

    data = yf.download(stock + ".NS", start=start_date, end=end_date)
    
    # Flatten column names in case multi-index appears
    data.columns = [col[0] if isinstance(col, tuple) else col for col in data.columns]
    
    # Keep only relevant columns
    data = data[['Open', 'High', 'Low', 'Close', 'Volume']].copy()
    
    # Add stock id and reset index
    data['stock_id'] = stock
    data.reset_index(inplace=True)  # index → Date
    
    all_data.append(data)

# Combine all stocks into one DataFrame
all_data = pd.concat(all_data, ignore_index=True)

# Reorder columns
all_data = all_data[['Date', 'stock_id', 'Open', 'High', 'Low', 'Close', 'Volume']]

print(all_data.head())
print("Final shape:", all_data.shape)
```
